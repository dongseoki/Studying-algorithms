```python
def test(n, max_point, caseIntList, onePointMoreList):
    if sum(caseIntList)> n:
        return max_point, []
    lianInfoList = [ pair[0] * pair[1]  for pair in zip(onePointMoreList, caseIntList)]
    lianPoint = calculatePoint(lianInfoList)
    if max_point < lianPoint:
        max_point = lianPoint
        max_point_case_list = lianInfoList
        return max_point, max_point_case_list
    else :
        return max_point, []
```
